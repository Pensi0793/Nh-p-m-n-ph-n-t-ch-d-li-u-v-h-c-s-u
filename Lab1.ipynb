```python
import pandas as pd
import numpy as np
```

```python
#huongdan1
df = pd.read_csv('dulieuxettuyendaihoc.csv', index_col="STT")
df.head(10)
```

```python
#huongdan2
df.tail(10)
```

```python
#huongdan4
df['DT'].fillna(0, inplace=True)
df
```

```python
#huongdan5
df["T1"].fillna(df["T1"].mean(), inplace=True)
df
```

```python
#cau 6
diem_cols = ['T1','L1','H1','S1','V1','X1','D1','N1',
             'T2','L2','H2','S2','V2','X2','D2','N2',
             'T6','L6','H6','S6','V6','X6','D6','N6']

for col in diem_cols:
    df[col].fillna(df[col].mean(), inplace=True)
df.isnull().sum()
```

```python
#cau7
df["TBM1"] = (df["T1"]*2 + df["L1"] + df["H1"] + df["S1"] +
              df["V1"]*2 + df["X1"] + df["D1"] + df["N1"]) / 10

df["TBM2"] = (df["T2"]*2 + df["L2"] + df["H2"] + df["S2"] +
              df["V2"]*2 + df["X2"] + df["D2"] + df["N2"]) / 10

df["TBM3"] = (df["T6"]*2 + df["L6"] + df["H6"] + df["S6"] +
              df["V6"]*2 + df["X6"] + df["D6"] + df["N6"]) / 10
df[["TBM1","TBM2","TBM3"]].head(10)
```

```python
#8
def xeploai(tbm):
    if tbm < 5:
        return "Y"
    elif tbm < 6.5:
        return "TB"
    elif tbm < 8:
        return "K"
    elif tbm < 9:
        return "G"
    else:
        return "XS"

df["XL1"] = df["TBM1"].apply(xeploai)
df["XL2"] = df["TBM2"].apply(xeploai)
df["XL3"] = df["TBM3"].apply(xeploai)
df[["TBM1","XL1","TBM2","XL2","TBM3","XL3"]].head()
```

```python
#cau9
df["US_TBM1"] = (df["TBM1"] - df["TBM1"].min()) / \
                (df["TBM1"].max() - df["TBM1"].min()) * 4

df["US_TBM2"] = (df["TBM2"] - df["TBM2"].min()) / \
                (df["TBM2"].max() - df["TBM2"].min()) * 4

df["US_TBM3"] = (df["TBM3"] - df["TBM3"].min()) / \
                (df["TBM3"].max() - df["TBM3"].min()) * 4
df[["US_TBM1","US_TBM2","US_TBM3"]].head()
```

```python
#cau10
def kqxt(row):
    if row["KT"] in ["A", "A1"]:
        diem = (row["DH1"]*2 + row["DH2"] + row["DH3"]) / 4

    elif row["KT"] == "B":
        diem = (row["DH1"] + row["DH2"]*2 + row["DH3"]) / 4

    else:
        diem = (row["DH1"] + row["DH2"] + row["DH3"]) / 3

    return 1 if diem >= 5 else 0

df["KQXT"] = df.apply(kqxt, axis=1)
df[["KT","DH1","DH2","DH3","KQXT"]].head()
```
