```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import string

print("\n=== BÀI 4: LOẠI BỎ STOPWORD TỪ VĂN BẢN ===")

van_ban = "This is a sample text for removing stopwords. We will remove all common stopwords from this text."
print(f"Văn bản gốc:\n{van_ban}")

# Tokenize và loại bỏ stopword
tokens = word_tokenize(van_ban.lower())
stop_words = set(stopwords.words('english'))
filtered_tokens = [word for word in tokens if word not in stop_words and word not in string.punctuation]

print(f"\nVăn bản sau khi loại bỏ stopword:\n{' '.join(filtered_tokens)}")
print(f"Tokens gốc: {len(tokens)} từ")
print(f"Tokens sau khi loại bỏ stopword: {len(filtered_tokens)} từ")
```

```python
from nltk.corpus import stopwords

print("\n=== BÀI 5: BỎ QUA STOPWORD TỪ DANH SÁCH ===")

def bo_qua_stopword(danh_sach_tu, ngon_ngu='english'):
    stop_words = set(stopwords.words(ngon_ngu))
    filtered_list = [word for word in danh_sach_tu if word.lower() not in stop_words]
    return filtered_list

# Ví dụ
danh_sach_tu = ['I', 'love', 'you', 'and', 'he', 'she', 'it', 'we', 'they']
print(f"Danh sách gốc: {danh_sach_tu}")
print(f"Danh sách sau khi bỏ stopword: {bo_qua_stopword(danh_sach_tu)}")
```

```python
from nltk.corpus import wordnet as wn

print("\n=== BÀI 6: ĐỊNH NGHĨA VÀ VÍ DỤ TỪ WORDNET ===")

def tim_dinh_nghia_vi_du(tu):
    print(f"\nTìm kiếm từ: '{tu}'")
    synsets = wn.synsets(tu)

    if not synsets:
        print(f"Không tìm thấy từ '{tu}' trong WordNet")
        return

    for i, synset in enumerate(synsets[:3], 1):  # Chỉ lấy 3 kết quả đầu
        print(f"\nĐịnh nghĩa {i}: {synset.definition()}")
        print(f"Ví dụ: {synset.examples() if synset.examples() else 'Không có ví dụ'}")

tim_dinh_nghia_vi_du('book')
tim_dinh_nghia_vi_du('computer')
```

```python
from nltk.corpus import wordnet as wn

print("\n=== BÀI 7: TỪ ĐỒNG NGHĨA VÀ TRÁI NGHĨA ===")

def tim_dong_nghia_trai_nghia(tu):
    print(f"\nTừ: '{tu}'")
    synonyms = set()
    antonyms = set()

    for syn in wn.synsets(tu):
        for lemma in syn.lemmas():
            synonyms.add(lemma.name())
            if lemma.antonyms():
                antonyms.add(lemma.antonyms()[0].name())

    print(f"Từ đồng nghĩa: {list(synonyms)[:10]}")
    print(f"Từ trái nghĩa: {list(antonyms)}")

tim_dong_nghia_trai_nghia('good')
tim_dong_nghia_trai_nghia('happy')
```

```python
import nltk
from nltk.tag import map_tag

print("\n=== BÀI 8: TỔNG QUAN VỀ BỘ TAG ===")

def hien_thi_tag_info():
    print("Tổng quan về các bộ tag:")
    print("-" * 50)

    # Tagset của Brown Corpus
    brown_tags = nltk.corpus.brown.tagged_words(tagset='universal')
    print(f"Số lượng tag trong Universal Tagset: {len(set(tag for word, tag in brown_tags))}")

    # Hiển thị các tag phổ biến
    print("\nCác tag trong Universal Tagset:")
    universal_tags = ['ADJ', 'ADP', 'ADV', 'CONJ', 'DET', 'NOUN', 'NUM', 'PRT', 'PRON', 'VERB', 'X', '.']
    for tag in sorted(universal_tags):
        count = sum(1 for word, t in brown_tags if t == tag)
        print(f"  {tag}: {count} từ")

    # Hiển thị chi tiết một tag cụ thể
    print("\nChi tiết tag 'NOUN' trong Universal Tagset:")
    print("  NOUN: Danh từ (noun)")
    print("  Ví dụ: dog, cat, book, computer")

    # Sử dụng biểu thức chính quy để tìm tag
    import re
    print("\nTìm các tag chứa 'NN' (danh từ trong Penn Treebank):")
    from nltk.corpus import treebank
    treebank_tags = set(tag for word, tag in treebank.tagged_words())
    nn_tags = [tag for tag in treebank_tags if re.search(r'NN', tag)]
    print(f"  Các tag chứa 'NN': {sorted(nn_tags)}")

hien_thi_tag_info()
```

```python
from nltk.corpus import wordnet as wn

print("\n=== BÀI 9: SO SÁNH SỰ GIỐNG NHAU CỦA HAI DANH TỪ ===")

def so_sanh_danh_tu(tu1, tu2):
    print(f"\nSo sánh danh từ: '{tu1}' và '{tu2}'")

    # Lấy synset đầu tiên của mỗi từ
    syn1 = wn.synsets(tu1, pos=wn.NOUN)
    syn2 = wn.synsets(tu2, pos=wn.NOUN)

    if not syn1:
        print(f"Không tìm thấy danh từ '{tu1}'")
        return
    if not syn2:
        print(f"Không tìm thấy danh từ '{tu2}'")
        return

    # Tính độ tương đồng
    sim = syn1[0].wup_similarity(syn2[0])
    if sim:
        print(f"Độ tương đồng (Wu-Palmer): {sim:.4f}")
    else:
        print("Không thể tính độ tương đồng")

so_sanh_danh_tu('dog', 'cat')
so_sanh_danh_tu('car', 'vehicle')
so_sanh_danh_tu('apple', 'orange')
```

```python
from nltk.corpus import wordnet as wn

print("\n=== BÀI 10: SO SÁNH SỰ GIỐNG NHAU CỦA HAI ĐỘNG TỪ ===")

def so_sanh_dong_tu(tu1, tu2):
    print(f"\nSo sánh động từ: '{tu1}' và '{tu2}'")

    # Lấy synset đầu tiên của mỗi từ
    syn1 = wn.synsets(tu1, pos=wn.VERB)
    syn2 = wn.synsets(tu2, pos=wn.VERB)

    if not syn1:
        print(f"Không tìm thấy động từ '{tu1}'")
        return
    if not syn2:
        print(f"Không tìm thấy động từ '{tu2}'")
        return

    # Tính độ tương đồng
    sim = syn1[0].path_similarity(syn2[0])
    if sim:
        print(f"Độ tương đồng (Path): {sim:.4f}")
    else:
        print("Không thể tính độ tương đồng")

so_sanh_dong_tu('run', 'walk')
so_sanh_dong_tu('eat', 'sleep')
so_sanh_dong_tu('run', 'sprint')
```

```python
from nltk.corpus import names
import random

print("\n=== BÀI 12: 15 KẾT HỢP NGẪU NHIÊN ĐƯỢC GẮN NHÃN ===")

nam_names = names.words('male.txt')
nu_names = names.words('female.txt')

# Tạo danh sách tên đã được gắn nhãn
ten_duoc_gan_nhan = [(name, 'male') for name in nam_names] + [(name, 'female') for name in nu_names]
random.seed(42)  # Để kết quả có thể tái tạo
mau_ngau_nhien = random.sample(ten_duoc_gan_nhan, 15)

print("15 kết hợp tên và nhãn ngẫu nhiên:")
for i, (name, label) in enumerate(mau_ngau_nhien, 1):
    print(f"  {i}. Tên: {name} - Nhãn: {label}")
```

```python
from nltk.corpus import names

print("\n=== BÀI 13: TRÍCH XUẤT KÝ TỰ CUỐI CÙNG CỦA TÊN ===")

nam_names = names.words('male.txt')
nu_names = names.words('female.txt')

# Tạo mảng mới với ký tự cuối cùng và nhãn
def trich_xuat_ky_tu_cuoi():
    ket_qua = []
    for name in nam_names[:50]:  # Lấy 50 tên nam để minh họa
        if len(name) > 0:
            ket_qua.append((name[-1], 'male'))
    for name in nu_names[:50]:  # Lấy 50 tên nữ để minh họa
        if len(name) > 0:
            ket_qua.append((name[-1], 'female'))
    return ket_qua

mang_moi = trich_xuat_ky_tu_cuoi()
print("Mảng mới (ký tự cuối, nhãn) - 20 phần tử đầu:")
for i, (char, label) in enumerate(mang_moi[:20], 1):
    print(f"  {i}. '{char}' - {label}")

# Thống kê ký tự cuối phổ biến
print("\nThống kê 10 ký tự cuối phổ biến nhất trong tên nam:")
male_end_chars = {}
for name in nam_names:
    char = name[-1]
    male_end_chars[char] = male_end_chars.get(char, 0) + 1
sorted_male = sorted(male_end_chars.items(), key=lambda x: x[1], reverse=True)[:10]
for char, count in sorted_male:
    print(f"  '{char}': {count} tên")

print("\nThống kê 10 ký tự cuối phổ biến nhất trong tên nữ:")
female_end_chars = {}
for name in nu_names:
    char = name[-1]
    female_end_chars[char] = female_end_chars.get(char, 0) + 1
sorted_female = sorted(female_end_chars.items(), key=lambda x: x[1], reverse=True)[:10]
for char, count in sorted_female:
    print(f"  '{char}': {count} tên")
```
