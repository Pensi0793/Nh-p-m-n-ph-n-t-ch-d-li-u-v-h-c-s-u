```python
#Phan1
#Cau1
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from datetime import datetime
df = pd.read_excel('ecommerce_sales_data.xlsx')
```

```python
#cau2
print("7 DÒNG ĐẦU TIÊN CỦA DỮ LIỆU:")
print(df.head(7))
print("5 DÒNG CUỐI CÙNG CỦA DỮ LIỆU:")
print(df.tail(5))
```

```python
#cau3
print("\n📊 THÔNG TIN TỔNG QUAN VỀ DỮ LIỆU:")
print(f"Số dòng: {df.shape[0]}")
print(f"Số cột: {df.shape[1]}")
print("\nKiểu dữ liệu của từng cột:")
print(df.dtypes)
```

```python
#cau4
print("\n📊 THỐNG KÊ MÔ TẢ CHO CÁC CỘT SỐ:")
print(df.describe())
```

```python
#Phan2
#cau1
print("\n🔍 KIỂM TRA GIÁ TRỊ BỊ THIẾU:")
print(df.isnull().sum())
```

```python
#cau2
print("\n🔄 XỬ LÝ GIÁ TRỊ THIẾU Ở CỘT DISCOUNT:")
df['Discount'] = df['Discount'].fillna(0)
print(f"Số giá trị thiếu còn lại: {df['Discount'].isnull().sum()}")
```

```python
#cau3
print("\n🔄 XỬ LÝ GIÁ TRỊ THIẾU Ở CỘT CUSTOMERID:")
df['CustomerID'] = df['CustomerID'].fillna('GUEST')
print(f"Số giá trị thiếu còn lại: {df['CustomerID'].isnull().sum()}")
```

```python
#cau4
print("\n✅ KIỂM TRA LẠI SAU KHI XỬ LÝ:")
print(df.isnull().sum())
```

```python
#CÂU 5: Chuyển đổi kiểu dữ liệu của cột Date

print(f"Kiểu dữ liệu cũ: {df['Date'].dtype}")
print(f"Ví dụ 5 giá trị đầu tiên:")
print(df['Date'].head())

# Chuyển đổi sang datetime
df['Date'] = pd.to_datetime(df['Date'])

print(f"\nKiểu dữ liệu mới: {df['Date'].dtype}")
print(f"Ví dụ 5 giá trị sau chuyển đổi:")
print(df['Date'].head())
print(f"\nNgày nhỏ nhất: {df['Date'].min()}")
print(f"Ngày lớn nhất: {df['Date'].max()}")
```

```python
# CÂU 6: Kiểm tra và loại bỏ dữ liệu trùng lặp
# Kiểm tra số dòng trùng lặp
duplicate_count = df.duplicated().sum()
print(f"Số dòng trùng lặp: {duplicate_count}")

if duplicate_count > 0:
    print("\nVí dụ về dòng trùng lặp:")
    print(df[df.duplicated(keep=False)].head())

    # Xóa dòng trùng lặp
    df = df.drop_duplicates()
    print(f"\nĐã xóa {duplicate_count} dòng trùng lặp")
    print(f"Số dòng còn lại: {len(df)}")
else:
    print("\nKhông có dữ liệu trùng lặp!")
```

```python
# PHẦN 3
#CÂU 1: Tạo cột Revenue
# Tạo cột Revenue
df['Revenue'] = df['Quantity'] * df['UnitPrice'] * (1 - df['Discount'])

print("Đã tạo cột Revenue thành công!")
print("\nThống kê cơ bản:")
print(f"  - Revenue trung bình: ${df['Revenue'].mean():,.2f}")
print(f"  - Revenue thấp nhất: ${df['Revenue'].min():,.2f}")
print(f"  - Revenue cao nhất: ${df['Revenue'].max():,.2f}")
print(f"  - Tổng Revenue: ${df['Revenue'].sum():,.2f}")

print("\nVí dụ 5 đơn hàng đầu tiên:")
print(df[['OrderID', 'Product_Name', 'Quantity', 'UnitPrice', 'Discount', 'Revenue']].head())
```

```python

#CÂU 2
# Trích xuất năm
df['Year'] = df['Date'].dt.year

print("Đã tạo cột Year thành công!")
print(f"\nCác năm trong dữ liệu: {sorted(df['Year'].unique().tolist())}")

print("\nThống kê theo năm:")
year_stats = df.groupby('Year').agg({
    'OrderID': 'count',
    'Revenue': 'sum'
}).rename(columns={'OrderID': 'Số đơn hàng'})

for year, row in year_stats.iterrows():
    print(f"  Năm {year}: {row['Số đơn hàng']} đơn hàng, doanh thu: ${row['Revenue']:,.2f}")
```

```python
#CÂU 3
# Trích xuất tháng
df['Month'] = df['Date'].dt.month

print("Đã tạo cột Month thành công!")
print(f"\nCác tháng trong dữ liệu: {sorted(df['Month'].unique().tolist())}")

print("\nThống kê theo tháng:")
month_stats = df.groupby('Month').agg({
    'OrderID': 'count',
    'Revenue': 'sum'
}).rename(columns={'OrderID': 'Số đơn hàng'})

for month, row in month_stats.iterrows():
    print(f"  Tháng {month}: {row['Số đơn hàng']} đơn hàng, doanh thu: ${row['Revenue']:,.2f}")

# Thêm tên tháng để dễ đọc
month_names = {
    1: 'Tháng 1', 2: 'Tháng 2', 3: 'Tháng 3', 4: 'Tháng 4',
    5: 'Tháng 5', 6: 'Tháng 6', 7: 'Tháng 7', 8: 'Tháng 8',
    9: 'Tháng 9', 10: 'Tháng 10', 11: 'Tháng 11', 12: 'Tháng 12'
}
df['Month_Name'] = df['Month'].map(month_names)
print("\nĐã thêm cột Month_Name!")
```

```python
# PHẦN 3 - CÂU 4
# Trích xuất ngày trong tuần
df['DayOfWeek'] = df['Date'].dt.day_name()

print("Đã tạo cột DayOfWeek thành công!")
print(f"\nCác ngày trong tuần: {sorted(df['DayOfWeek'].unique().tolist())}")

print("\nThống kê theo ngày trong tuần:")
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
day_stats = df.groupby('DayOfWeek').agg({
    'OrderID': 'count',
    'Revenue': 'sum'
}).rename(columns={'OrderID': 'Số đơn hàng'})

for day in day_order:
    if day in day_stats.index:
        row = day_stats.loc[day]
        print(f"  {day}: {row['Số đơn hàng']} đơn hàng, doanh thu: ${row['Revenue']:,.2f}")
```

```python
# PHẦN 3 - CÂU 5:

# Định nghĩa hàm phân loại
def categorize_price(price):
    if price >= 150:
        return 'High'
    elif price >= 50:
        return 'Medium'
    else:
        return 'Low'

# Tạo cột Price_Segment
df['Price_Segment'] = df['UnitPrice'].apply(categorize_price)

print("Đã tạo cột Price_Segment thành công!")
print("\nPhân phối theo phân khúc giá:")
print(df['Price_Segment'].value_counts())

print("\nChi tiết từng phân khúc:")
segment_stats = df.groupby('Price_Segment').agg({
    'UnitPrice': ['count', 'mean', 'min', 'max'],
    'Revenue': 'sum'
})
segment_stats.columns = ['Số lượng', 'Giá TB', 'Giá Min', 'Giá Max', 'Tổng doanh thu']

for segment in ['High', 'Medium', 'Low']:
    if segment in segment_stats.index:
        row = segment_stats.loc[segment]
        print(f"\n  {segment}:")
        print(f"    - Số lượng sản phẩm: {row['Số lượng']}")
        print(f"    - Giá trung bình: ${row['Giá TB']:,.2f}")
        print(f"    - Giá từ ${row['Giá Min']:,.2f} đến ${row['Giá Max']:,.2f}")
        print(f"    - Tổng doanh thu: ${row['Tổng doanh thu']:,.2f}")
```

```python
# PHẦN 4 - CÂU 1: Lọc đơn hàng khu vực North có Revenue > 300

north_high = df[(df['Region'] == 'North') & (df['Revenue'] > 300)]

if len(north_high) > 0:
    print(f"Tìm thấy {len(north_high)} đơn hàng")
    print("\nChi tiết đơn hàng:")
    print(north_high[['OrderID', 'Product_Name', 'Revenue', 'Date', 'Region']].to_string(index=False))
else:
    print("Không tìm thấy đơn hàng nào!")
```

```python
# PHẦN 4 - CÂU 2: Đếm đơn hàng Electronics không giảm giá

electronics_no_discount = df[(df['Product_Category'] == 'Electronics') & (df['Discount'] == 0)]

print(f"Số đơn hàng Electronics không được giảm giá: {len(electronics_no_discount)}")
if len(electronics_no_discount) > 0:
    print(f"Tổng doanh thu: ${electronics_no_discount['Revenue'].sum():,.2f}")
    print("\nVí dụ 5 đơn hàng đầu tiên:")
    print(electronics_no_discount[['OrderID', 'Product_Name', 'Quantity', 'UnitPrice', 'Revenue']].head())
```

```python
# PHẦN 4 - CÂU 3: Lọc đơn hàng từ tháng 3 đến tháng 6

march_to_june = df[(df['Month'] >= 3) & (df['Month'] <= 6)]

print(f"Số đơn hàng trong khoảng tháng 3-6: {len(march_to_june)}")
print(f"Tổng doanh thu: ${march_to_june['Revenue'].sum():,.2f}")
print(f"Doanh thu trung bình/đơn: ${march_to_june['Revenue'].mean():,.2f}")

print("\n5 đơn hàng đầu tiên:")
print(march_to_june[['OrderID', 'Product_Name', 'Revenue', 'Date', 'Month']].head())
```

```python
# PHẦN 4 - CÂU 4: Top 10 đơn hàng doanh thu cao nhất

top_10 = df.nlargest(10, 'Revenue')

print("Top 10 đơn hàng có doanh thu cao nhất:")
print("\n" + "-"*80)
for i, (idx, row) in enumerate(top_10.iterrows(), 1):
    print(f"{i:2d}. OrderID: {row['OrderID']:8s} | {row['Product_Name'][:30]:30s} | ${row['Revenue']:10,.2f} | {row['Date'].strftime('%Y-%m-%d')}")
print("-"*80)

print(f"\nTổng doanh thu của 10 đơn hàng này: ${top_10['Revenue'].sum():,.2f}")
```

```python
# PHẦN 4 - CÂU 5: Đơn hàng có số lượng mua nhiều nhất

max_quantity = df['Quantity'].max()
max_qty_orders = df[df['Quantity'] == max_quantity]

print(f"Số lượng mua nhiều nhất: {max_quantity} sản phẩm")
print(f"Số đơn hàng có số lượng này: {len(max_qty_orders)}")

print("\nChi tiết đơn hàng:")
for i, (idx, row) in enumerate(max_qty_orders.iterrows(), 1):
    print(f"\n  Đơn hàng {i}:")
    print(f"    - OrderID: {row['OrderID']}")
    print(f"    - Tên sản phẩm: {row['Product_Name']}")
    print(f"    - Danh mục: {row['Product_Category']}")
    print(f"    - Số lượng: {row['Quantity']}")
    print(f"    - Đơn giá: ${row['UnitPrice']:,.2f}")
    print(f"    - Giảm giá: {row['Discount']*100:.0f}%")
    print(f"    - Doanh thu: ${row['Revenue']:,.2f}")
    print(f"    - Ngày mua: {row['Date'].strftime('%Y-%m-%d')}")
    print(f"    - Khu vực: {row['Region']}")
```

```python
# PHẦN 5 - CÂU 1: Tổng doanh thu theo khu vực

region_revenue = df.groupby('Region')['Revenue'].sum().sort_values(ascending=False)

print("Bảng tổng hợp doanh thu theo khu vực:")
print("\n" + "-"*50)
print(f"{'Khu vực':<15} {'Doanh thu':>20} {'Tỷ lệ %':>15}")
print("-"*50)

total_revenue = region_revenue.sum()
for region, revenue in region_revenue.items():
    percentage = (revenue / total_revenue) * 100
    print(f"{region:<15} ${revenue:>19,.2f} {percentage:>14.1f}%")
print("-"*50)
print(f"{'TỔNG':<15} ${total_revenue:>19,.2f} {'100.0%':>15}")
print("-"*50)

print(f"\n🏆 Khu vực có doanh thu cao nhất: {region_revenue.index[0]} (${region_revenue.iloc[0]:,.2f})")
print(f"📉 Khu vực có doanh thu thấp nhất: {region_revenue.index[-1]} (${region_revenue.iloc[-1]:,.2f})")
```

```python
# PHẦN 5 - CÂU 2: Trung bình Quantity và UnitPrice theo danh mục

category_stats = df.groupby('Product_Category').agg({
    'Quantity': ['mean', 'sum'],
    'UnitPrice': ['mean', 'min', 'max'],
    'Revenue': 'sum'
})
category_stats.columns = ['SL Trung bình', 'Tổng SL', 'Giá TB', 'Giá Min', 'Giá Max', 'Tổng doanh thu']
category_stats = category_stats.round(2)

print("Thống kê theo danh mục sản phẩm:")
print("\n" + "="*100)
for category, row in category_stats.iterrows():
    print(f"\n📁 {category}:")
    print(f"   • Số lượng trung bình/đơn: {row['SL Trung bình']:.2f} sản phẩm")
    print(f"   • Tổng số lượng bán ra: {row['Tổng SL']:.0f} sản phẩm")
    print(f"   • Giá trung bình: ${row['Giá TB']:,.2f}")
    print(f"   • Giá từ ${row['Giá Min']:,.2f} đến ${row['Giá Max']:,.2f}")
    print(f"   • Tổng doanh thu: ${row['Tổng doanh thu']:,.2f}")
print("="*100)
```

```python
# PHẦN 5 - CÂU 3: Số lượng đơn hàng theo tháng

monthly_orders = df.groupby('Month').size().sort_values(ascending=False)

print("Bảng thống kê đơn hàng theo tháng:")
print("\n" + "-"*50)
print(f"{'Tháng':<10} {'Số đơn hàng':>15} {'Tỷ lệ %':>15} {'Doanh thu':>15}")
print("-"*50)

total_orders = monthly_orders.sum()
monthly_revenue = df.groupby('Month')['Revenue'].sum()

for month in range(1, 13):
    if month in monthly_orders.index:
        count = monthly_orders[month]
        revenue = monthly_revenue[month]
        percentage = (count / total_orders) * 100
        print(f"{month:>2}:       {count:>10}   {percentage:>13.1f}%   ${revenue:>14,.2f}")
print("-"*50)
print(f"{'TỔNG':<10} {total_orders:>15}   {'100.0%':>15}   ${monthly_revenue.sum():>14,.2f}")
print("-"*50)

print(f"\n📊 Tháng có nhiều đơn hàng nhất: Tháng {monthly_orders.index[0]} ({monthly_orders.iloc[0]} đơn)")
print(f"📊 Tháng có ít đơn hàng nhất: Tháng {monthly_orders.index[-1]} ({monthly_orders.iloc[-1]} đơn)")
```

```python
# PHẦN 5 - CÂU 4: Top 3 khách hàng chi tiêu nhiều nhất
# Loại bỏ khách hàng GUEST
non_guest = df[df['CustomerID'] != 'GUEST']

if len(non_guest) > 0:
    top_customers = non_guest.groupby('CustomerID')['Revenue'].sum().sort_values(ascending=False).head(3)

    print("Top 3 khách hàng chi tiêu nhiều nhất (không tính GUEST):")
    print("\n" + "-"*70)
    print(f"{'Hạng':<6} {'Mã khách hàng':<15} {'Tổng chi tiêu':>20} {'Số đơn hàng':>15} {'TB/đơn':>15}")
    print("-"*70)

    for rank, (customer_id, total_revenue) in enumerate(top_customers.items(), 1):
        # Đếm số đơn hàng của khách hàng
        order_count = len(non_guest[non_guest['CustomerID'] == customer_id])
        avg_per_order = total_revenue / order_count

        print(f"{rank:<6} {customer_id:<15} ${total_revenue:>19,.2f} {order_count:>14}   ${avg_per_order:>14,.2f}")
    print("-"*70)

    # Thông tin thêm về top 1
    top1 = top_customers.index[0]
    top1_details = non_guest[non_guest['CustomerID'] == top1]
    print(f"\n🏆 Khách hàng chi tiêu nhiều nhất: {top1}")
    print(f"   • Tổng chi tiêu: ${top_customers.iloc[0]:,.2f}")
    print(f"   • Số đơn hàng: {len(top1_details)}")
    print(f"   • Trung bình/đơn: ${top_customers.iloc[0]/len(top1_details):,.2f}")
    print(f"   • Danh mục ưa thích: {top1_details['Product_Category'].mode().iloc[0]}")
else:
    print("Không có khách hàng nào không phải GUEST!")
```

```python
# PHẦN 6 - CÂU 1: Biểu đồ cột dọc doanh thu theo danh mục

# Thiết lập kích thước biểu đồ
plt.figure(figsize=(12, 7))

# Tính tổng doanh thu theo danh mục và sắp xếp giảm dần
category_revenue = df.groupby('Product_Category')['Revenue'].sum().sort_values(ascending=False)

# Tạo màu sắc cho biểu đồ
colors = sns.color_palette("viridis", len(category_revenue))

# Vẽ biểu đồ cột dọc
bars = plt.bar(category_revenue.index, category_revenue.values,
               color=colors, edgecolor='black', linewidth=0.5, alpha=0.8)

# Thêm tiêu đề và nhãn
plt.title('TỔNG DOANH THU THEO DANH MỤC SẢN PHẨM', fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Danh mục sản phẩm', fontsize=13, fontweight='bold')
plt.ylabel('Doanh thu (USD)', fontsize=13, fontweight='bold')

# Thêm grid
plt.grid(axis='y', alpha=0.3, linestyle='--')

# Thêm giá trị trên đỉnh cột
for i, bar in enumerate(bars):
    height = bar.get_height()
    # Thêm dấu $ và định dạng số
    plt.text(bar.get_x() + bar.get_width()/2, height + height*0.01,
             f'${height:,.0f}',
             ha='center', va='bottom', fontsize=11, fontweight='bold',
             rotation=0, color='darkblue')

# Xoay nhãn trục X nếu cần
plt.xticks(rotation=30, ha='right', fontsize=11)

# Tự động điều chỉnh layout
plt.tight_layout()

# Hiển thị biểu đồ
plt.show()

# In thông tin thống kê kèm theo
print("\n📊 BẢNG TỔNG HỢP DOANH THU THEO DANH MỤC:")
print("-" * 70)
print(f"{'STT':<6} {'Danh mục':<25} {'Doanh thu':>20} {'Tỷ lệ %':>15}")
print("-" * 70)

total_revenue = category_revenue.sum()
for i, (category, revenue) in enumerate(category_revenue.items(), 1):
    percentage = (revenue / total_revenue) * 100
    print(f"{i:<6} {category:<25} ${revenue:>19,.2f} {percentage:>14.1f}%")

print("-" * 70)
print(f"{'TỔNG':<6} {'':<25} ${total_revenue:>19,.2f} {'100.0%':>15}")
print("-" * 70)

# In danh mục có doanh thu cao nhất và thấp nhất
print(f"\n🏆 Danh mục có doanh thu cao nhất: {category_revenue.index[0]} (${category_revenue.iloc[0]:,.2f})")
print(f"📉 Danh mục có doanh thu thấp nhất: {category_revenue.index[-1]} (${category_revenue.iloc[-1]:,.2f})")
```

```python
# PHẦN 6 - CÂU 2: Biểu đồ đường doanh thu theo tháng

plt.figure(figsize=(12, 6))

# Tính tổng doanh thu theo tháng
monthly_revenue = df.groupby('Month')['Revenue'].sum()

# Vẽ biểu đồ đường
plt.plot(monthly_revenue.index, monthly_revenue.values,
         marker='o', linewidth=2.5, markersize=10, color='#FF6B6B',
         markerfacecolor='white', markeredgewidth=2, markeredgecolor='#FF6B6B')

# Thêm tiêu đề và nhãn
plt.title('BIẾN ĐỘNG DOANH THU THEO THÁNG', fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Tháng', fontsize=13, fontweight='bold')
plt.ylabel('Tổng doanh thu (USD)', fontsize=13, fontweight='bold')

# Thêm grid
plt.grid(True, alpha=0.3, linestyle='--')
plt.xticks(range(1, 13), fontsize=11)
plt.yticks(fontsize=11)

# Thêm giá trị trên biểu đồ
for month, revenue in monthly_revenue.items():
    plt.text(month, revenue + revenue*0.02, f'${revenue:,.0f}',
             ha='center', va='bottom', fontsize=10, fontweight='bold',
             color='#FF6B6B')

# Tự động điều chỉnh layout
plt.tight_layout()
plt.show()

# In thống kê
print("\n📊 THỐNG KÊ DOANH THU THEO THÁNG:")
print("-" * 80)
print(f"{'Tháng':<10} {'Doanh thu':>20} {'Tăng/Giảm so với tháng trước':>35}")
print("-" * 80)

prev_revenue = None
for month, revenue in monthly_revenue.items():
    if prev_revenue is not None:
        change = ((revenue - prev_revenue) / prev_revenue) * 100
        if change > 0:
            change_str = f"+{change:.1f}% ↑"
            # Màu xanh cho tăng
            print(f"Tháng {month:<2}    ${revenue:>18,.2f}    {change_str:>35}")
        else:
            change_str = f"{change:.1f}% ↓"
            # Màu đỏ cho giảm
            print(f"Tháng {month:<2}    ${revenue:>18,.2f}    {change_str:>35}")
    else:
        print(f"Tháng {month:<2}    ${revenue:>18,.2f}    {'-':>35}")
    prev_revenue = revenue
print("-" * 80)

# Tìm tháng có doanh thu cao nhất và thấp nhất
max_month = monthly_revenue.idxmax()
min_month = monthly_revenue.idxmin()
print(f"\n🏆 Tháng có doanh thu cao nhất: Tháng {max_month} (${monthly_revenue[max_month]:,.2f})")
print(f"📉 Tháng có doanh thu thấp nhất: Tháng {min_month} (${monthly_revenue[min_month]:,.2f})")
```
