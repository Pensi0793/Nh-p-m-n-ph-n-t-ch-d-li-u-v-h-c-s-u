```python
# Import thư viện cần thiết
import pandas as pd

# ====================== PROBLEM 1 ======================
# Thêm header vào dataframe để diễn giải dữ liệu
column_names = ["Id", "Name", "Age", "Weight", "m006", "m612", "m1218", "f006", "f0612", "f1218"]
# Đọc file dữ liệu
df = pd.read_csv("/content/patient_heart_rate.csv", names = column_names, header=None)
# Hiển thị một vài dòng dữ liệu đầu tiên ra màn hình
print("=== PROBLEM 1 ===")
print(df.head())
print("\n")

# ====================== PROBLEM 2 ======================
# Xử lý vấn đề một cột lưu hỗn hợp nhiều dữ liệu (tách Name thành Firstname và Lastname)
print("=== PROBLEM 2 ===")
df[['Firstname','Lastname']] = df['Name'].str.split(expand=True)
df = df.drop('Name', axis=1)
print(df.head())
print("\n")

# ====================== PROBLEM 3 ======================
# Xử lý cột Weight: chuyển từ lbs sang kgs
print("=== PROBLEM 3 ===")

for i in range(0,len(df['Weight'])):
    original_weight_str = str(df.loc[i, 'Weight']).strip().lower()

    processed_weight_value = None

    if original_weight_str.endswith("lbs"):
        numeric_part = original_weight_str.replace("lbs", "").strip()
        if numeric_part.replace('.', '', 1).isdigit():
            float_x = float(numeric_part)
            processed_weight_value = int(float_x / 2.2) # Convert lbs to kgs
    elif original_weight_str.endswith("kgs"):
        numeric_part = original_weight_str.replace("kgs", "").strip()
        if numeric_part.replace('.', '', 1).isdigit():
            float_x = float(numeric_part)
            processed_weight_value = int(float_x) # Already in kgs, just convert to int
    elif original_weight_str.replace('.', '', 1).isdigit() and original_weight_str != '':
        # Assume it's a number without unit, treat as kgs
        float_x = float(original_weight_str)
        processed_weight_value = int(float_x)

    if processed_weight_value is not None:
        df.loc[i, 'Weight'] = str(processed_weight_value) + "kgs"
    else:
        # Handle cases that couldn't be parsed (e.g., empty string, non-numeric)
        df.loc[i, 'Weight'] = None # Or pd.NA or an empty string, depending on desired outcome

print(df.head())
print("\n")

# ====================== PROBLEM 4 ======================
# Xóa các hàng hoàn toàn rỗng (NaN)
print("=== PROBLEM 4 ===")
df.dropna(how="all", inplace=True)
print(df.head())
print("\n")

# ====================== PROBLEM 5 ======================
# Xóa các hàng trùng lặp dựa trên các cột cụ thể
print("=== PROBLEM 5 ===")
df = df.drop_duplicates(subset=['Firstname', 'Lastname', 'Age', 'Weight'])
print(df.head())
print("\n")

# ====================== PROBLEM 6 ======================
# Xóa ký tự đặc biệt trong cột Firstname và Lastname (chỉ giữ chữ cái)
print("=== PROBLEM 6 ===")
df['Firstname'] = df['Firstname'].astype(str).str.replace(r'[^\x00-\x7F]+','', regex=True)
df['Lastname'] = df['Lastname'].astype(str).str.replace(r'[^\x00-\x7F]+','', regex=True)
print(df.head())
print("\n")

# ====================== PROBLEM 7 ======================
# Chuyển đổi dữ liệu từ dạng wide sang long (melt) và trích xuất thông tin Sex + Time
print("=== PROBLEM 7 ===")
# Melt: gộp các cột m006, m612, m1218, f006, f0612, f1218 thành 1 cột
df = pd.melt(df, id_vars=['Id', 'Age', 'Weight', 'Firstname', 'Lastname'],
             value_name="PulseRate", var_name="sex_and_time").sort_values(['Id', 'Age', 'Weight', 'Firstname', 'Lastname'])

# Trích xuất Sex, Hour Lower bound và Hour Upper bound
# tmp_df = df["sex_and_time"].str.extract(r"(\D+)(\d{2})?(\d{2})?", expand=True)

# Xử lý để lấy đúng định dạng
df["Sex"] = df["sex_and_time"].str.extract(r'([mf])')
df["hours_lower"] = df["sex_and_time"].str.extract(r'(\d{2})(?=\d{2})')
df["hours_upper"] = df["sex_and_time"].str.extract(r'(\d{2})$')

# Tạo cột Time
df["Time"] = df["hours_lower"] + "-" + df["hours_upper"]

# Bỏ các cột không cần thiết và xóa hàng NaN
df = df.drop(['sex_and_time', 'hours_lower', 'hours_upper'], axis=1)
df = df.dropna()

# Lưu ra file CSV
df.to_csv('outputcleanup.csv', index=False)

# Hiển thị kết quả cuối cùng
print(df.head())
print("\n")
print("=== HOÀN THÀNH! File đã được lưu thành 'outputcleanup.csv' ===")
```
